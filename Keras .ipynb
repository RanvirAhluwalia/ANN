```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('../DATA/fake_reg.csv')
```

```python
df.head()
```

```python
x = df[['feature1', 'feature2']].values
```

```python
y = df['price'].values
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test,y_train,y_test = train_test_split(x,y,test_size = 0.3,random_state = 100)
```

```python
x_train.shape
```

```python
y_train.shape
```

```python
x_test.shape
```

```python
y_test.shape
```

```python
import tensorflow as tf
```

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
scaler = MinMaxScaler()
```

```python
scaler.fit(x_train)
```

```python
x_train = scaler.transform(x_train)
```

```python
x_test = scaler.transform(x_test)
```

```python
x_train
```

```python
x_test
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense,Activation
```

```python
model = Sequential()
```

```python
model.add(Dense(4,activation = 'relu'))
```

```python
model.add(Dense(4,activation = 'relu'))
```

```python
model.add(Dense(4,activation = 'relu'))
```

```python
model.add(Dense(1))
```

```python
model.compile(optimizer = 'rmsprop', loss = 'mse')
```

```python
model.fit(x = x_train, y = y_train, epochs = 250)
```

```python
model.history.history
```

```python
df = pd.DataFrame(model.history.history)
```

```python
df.plot()
```

```python
test_predictions = model.predict(x_test)
```

```python
test_predictions
```

```python
y_test
```

```python
plt.scatter(y_test,test_predictions)
```

```python
df1 = pd.DataFrame(y_test)
```

```python
df1
```

```python
df1.columns = ['Real Values']
```

```python
df2 = pd.DataFrame(test_predictions)
```

```python
df2.columns = ['Model Predictions']
```

```python
df3 = pd.concat([df1,df2], axis = 1)
```

```python
df3
```

```python
df3['Error'] = df3['Real Values'] - df3['Model Predictions']
```

```python
df3
```

```python
sns.set_style('whitegrid')
sns.distplot(df3['Error'], bins = 30, kde = False)
```

```python
from sklearn.metrics import mean_absolute_error,mean_squared_error
```

```python
print(mean_absolute_error(df3['Real Values'], df3['Model Predictions']))
```

```python
print(mean_squared_error(df3['Real Values'], df3['Model Predictions']))
```

```python
print(np.sqrt(mean_squared_error(df3['Real Values'], df3['Model Predictions'])))
```
